```python
data.drop(columns=['DEPARTURE_STATION_CD', 'AIRLINE_CD', 'FLIGHT_TIME'], inplace=True)
data
```

```python
data['AIRCRAFT_MAKE'] = data['AIRCRAFT_TYPE'].apply(lambda x: x[0])
data[['AIRCRAFT_MAKE', 'AIRCRAFT_TYPE']]
```

```python
data.drop(columns=['AIRCRAFT_TYPE'], inplace=True)
data
```

```python
from datetime import date
data['DAY_TYPE'] = data['FLIGHT_DATE'].apply(date.weekday)
data['DAY_TYPE'] = np.where(data['DAY_TYPE']<5, "WEEKEND", "WEEKDAY")
data.drop(columns=['FLIGHT_DATE'], inplace=True)
data
```

```python
data['FLIGHT_NO'] = data['FLIGHT_NO'].apply(lambda x: x[2])
data
```

```python
data.columns
```

```python
HAUL_data = data.groupby(by=['HAUL'])

HAUL_data = HAUL_data[['FIRST_CLASS_SEATS', 'BUSINESS_CLASS_SEATS', 
        'ECONOMY_SEATS','TIER1_ELIGIBLE_PAX','TIER2_ELIGIBLE_PAX', 'TIER3_ELIGIBLE_PAX']].mean().sort_values(by=['HAUL'])

HAUL_data
```

```python
HAUL_data['Tier1%'] = np.where( HAUL_data['FIRST_CLASS_SEATS'] > 0,
    HAUL_data['TIER1_ELIGIBLE_PAX']/(HAUL_data['FIRST_CLASS_SEATS'] + 0.05*HAUL_data['BUSINESS_CLASS_SEATS']),
    HAUL_data['TIER1_ELIGIBLE_PAX']/(0.5*HAUL_data['BUSINESS_CLASS_SEATS'])
    )


HAUL_data['Tier2%'] = np.where( HAUL_data['FIRST_CLASS_SEATS'] > 0,
    HAUL_data['TIER2_ELIGIBLE_PAX']/(HAUL_data['FIRST_CLASS_SEATS'] + 0.3*HAUL_data['BUSINESS_CLASS_SEATS'] + 0.1*HAUL_data['ECONOMY_SEATS']),
    HAUL_data['TIER2_ELIGIBLE_PAX']/(0.8*HAUL_data['BUSINESS_CLASS_SEATS'] + 0.1*HAUL_data['ECONOMY_SEATS'])
    )

HAUL_data['Tier3%'] =  np.where( HAUL_data['FIRST_CLASS_SEATS'] > 0,
    HAUL_data['TIER3_ELIGIBLE_PAX']/(HAUL_data['FIRST_CLASS_SEATS'] + HAUL_data['BUSINESS_CLASS_SEATS'] + 0.1*HAUL_data['ECONOMY_SEATS']),
    HAUL_data['TIER3_ELIGIBLE_PAX']/(HAUL_data['BUSINESS_CLASS_SEATS'] + 0.3*HAUL_data['ECONOMY_SEATS'])
    )


HAUL_data['TOTAL%_ELIGIBLE'] = HAUL_data['Tier1%'] + HAUL_data['Tier2%'] + HAUL_data['Tier3%']

HAUL_data
```
